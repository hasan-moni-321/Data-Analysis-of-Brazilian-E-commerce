### Merge of orders and  order_item dataset

```python
both_merge = pd.merge(orders,order_item, on='order_id', how='left')
both_merge.head()
```

```python
both_merge['order_purchase_timestamp']=pd.to_datetime(both_merge['order_purchase_timestamp'])
both_merge['order_delivered_customer_date']=pd.to_datetime(both_merge['order_delivered_customer_date'])
```

```python
both_merge['weekday']=both_merge['order_purchase_timestamp'].dt.day_name()
both_merge['year']=both_merge['order_purchase_timestamp'].dt.year
both_merge['day']=both_merge['order_purchase_timestamp'].dt.day
both_merge['weekday'] = both_merge['order_purchase_timestamp'].dt.weekday
both_merge['month']=both_merge['order_purchase_timestamp'].dt.month
both_merge['hour']=both_merge['order_purchase_timestamp'].dt.hour
```

```python
both_merge.head()
```

```python
pd.set_option('float_format', '{:f}'.format)


Year_price = both_merge.groupby('year')['price'].sum().sort_values(ascending=False).reset_index()

value_year = sns.barplot(x=Year_price['year'], y=Year_price['price'])
value_year.set_ylabel('Total value in every year')
```

```python
#boxplot of every year's total price
plt.figure(figsize=(8,8))
ax=sns.boxplot(x='year',y='price',data=Year_price, palette=sns.set_palette(palette='viridis_r'))
ax.set_xlabel('Year')
ax.set_ylabel('Total Value')
ax.set_title('Box Plot of transactions over the year')
```

```python
trend_month=pd.DataFrame(both_merge.groupby('month').agg({'price':'mean'}).rename(columns={'price':'mean_transaction'})).reset_index()
trend_month
```

```python
import calendar

x1 = trend_month.month.tolist()
y1 = trend_month.mean_transaction.tolist()
mapp = {}
for m,v in zip(x1, y1):
    mapp[m] = v
xn = [calendar.month_abbr[int(x)] for x in sorted(x1)]
vn = [mapp[x] for x in sorted(x1)]

plt.figure(figsize=(10,7))
ax=sns.barplot(x=xn,y=vn, color='#ed5569')
ax.set_title("Average value of transaction per month")
ax.set_xlabel('Month')
ax.set_ylabel('Value')
```

### frequency of the orders during the hour over the day

```python
hour_day = both_merge.groupby(['weekday', 'hour']).agg({'order_id':'count'}).rename(columns={'order_id':'freq'}).reset_index()
hour_day.head()
```

```python
day_dict = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
hour_day['weekday'] = hour_day['weekday'].map(day_dict)
hour_day.head()
```

```python
### Sorting it so that the plot order is correct.
hour_day['weekday']=pd.Categorical(hour_day['weekday'],categories=['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],ordered=True)

#convert to pivot table
hour_day=hour_day.pivot('weekday','hour','freq')
hour_day
```

```python
#visualizing pivot table
plt.figure(figsize=(15,8))
ax=sns.heatmap(hour_day, annot=True, fmt="d",cmap="OrRd")
ax.set_xlabel("Hour")
ax.set_ylabel("Day")
ax.set_title("Heatmap of tranactions over the hour by day",size=10)
```

most of the payment less than 1000 

```python
# not taking 0 in payment_value column
payment=payment[payment['payment_value']!=0]

plt.figure(figsize=(10,8))
plt.subplot(221)

ax=sns.distplot(np.log(payment[payment.payment_type=='credit_card']['payment_value'])+1, color="red")
ax.set_xlabel("Log Transaction value (BRL)")
ax.set_ylabel("Frequency")
ax.set_title("Distribution plot for credit card transactions")

plt.subplot(222)
ax1=sns.distplot(np.log(payment[payment.payment_type=='boleto']['payment_value'])+1,color="red")
ax1.set_xlabel("Log Transaction value (BRL)")
ax1.set_ylabel("Frequency")
ax1.set_title("Distribution plot for boleto transactions")

plt.subplot(223)
ax2=sns.distplot(np.log(payment[payment.payment_type=='debit_card']['payment_value'])+1,color="red")
ax2.set_xlabel("Log Transaction value (BRL)")
ax2.set_ylabel("Frequency")
ax2.set_title("Distribution plot for debit card transactions")

plt.subplot(224)
ax3=sns.distplot(np.log(payment[payment.payment_type=='voucher']['payment_value'])+1,color="red")
ax3.set_xlabel("Log Transaction value (BRL)")
ax3.set_ylabel("Frequency")
ax3.set_title("Distribution plot for voucher transactions")

plt.subplots_adjust(wspace = 0.5, hspace = 0.5,
                    top = 1.3)

plt.show()
```



From the distribution & box plot we understand the following

The distribution for credit card type of transaction is nearly normal and from the boxplot it is seen that there are extreme outliers in this case.This means that for higher value of transactions ,people prefer to buy on credit and pay later. Transaction through boleto is multimodal and seems to be for lesser value of BRL and here too the transactions are dominated by outliers.There seems to be a significant difference between credit card and boleto type of transactions . Debit card and vouchers are not used much and there are two modes dominating the distribution plot .

### order_delivered_customer_date column

```python
#dropping null data
orders_without_nan = orders.dropna()
orders_without_nan.head(30)
```

```python
orders_without_nan = orders_without_nan.copy()

# converting into datetime of orders_without_nan column
orders_without_nan['order_delivered_customer_date'] = pd.to_datetime(orders_without_nan['order_delivered_customer_date'])
```

```python
orders_without_nan.dtypes
```

```python
#making seperate column for year and month
orders_without_nan['deliver_year'] = pd.DatetimeIndex(orders_without_nan['order_delivered_customer_date']).year
orders_without_nan['deliver_month'] = pd.DatetimeIndex(orders_without_nan['order_delivered_customer_date']).month

orders_without_nan['deliver_hour'] = pd.DatetimeIndex(orders_without_nan['order_delivered_customer_date']).hour
```

```python
orders_without_nan.head()
```

```python
#total number of deliver in every year
deliver_per_year = orders_without_nan.groupby('deliver_year')['deliver_year'].count()
deliver_per_year.plot(kind='bar')
```

```python
#total number of deliver in every month
plt.figure(figsize=(14,6))
deliver_per_year = orders_without_nan.groupby('deliver_month')['deliver_month'].count()
deliver_per_year.plot(kind='bar')
```

Most of the deliver of the product happend in August month

```python
#total number of deliver in every hour
plt.figure(figsize=(14,6))
deliver_per_year = orders_without_nan.groupby('deliver_hour')['deliver_hour'].count()
deliver_per_year.plot(kind='bar')
```

Most of deliver happened in evening time
